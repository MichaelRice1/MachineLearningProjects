```python
import pandas as pd


player_data = pd.read_csv("C:/Users/athen/Desktop/PersonalProjects/Learning/nba_project/data/data/players_final/players.csv")

```

```python

def column_ordering(player_data):

    player_data = player_data.sort_values(by = ["player", "date"])

    return player_data


def convert_minutes(mp):
        try:
            parts = mp.split(":")
            if len(parts) == 2:
                return int(parts[0]) + int(parts[1]) / 60
            else:
                return float('nan')  # or handle it as needed
        except Exception as e:
            return float('nan')  # or handle it as needed


#feature engineering
def horizons(player_data):

    unique_players = player_data["player"].unique()

    
    cols_to_convert = ['PTS', 'TRB', 'AST',]

    for col in cols_to_convert:
        player_data[col] = pd.to_numeric(player_data[col], errors='coerce')



    #the MP col stores minutes played as a string, in the format MM:SS, convert to float, unless the value is 0, keep it as a 0
    if player_data["MP"].dtype == 'object':
        # Convert 'MP' column to numeric minutes
        player_data["MP"] = player_data["MP"].apply(convert_minutes)
    
        
    #add points per 38 minutes
    if 'pts_per_38' not in player_data.columns:
        player_data["pts_per_38"] = round((player_data["PTS"] / player_data["MP"]) * 38, 2)
    
    if 'trb_per_38' not in player_data.columns:
        player_data["trb_per_38"] = round((player_data["TRB"] / player_data["MP"]) * 38, 2)
    
    if 'ast_per_38' not in player_data.columns:
        player_data["ast_per_38"] = round((player_data["AST"] / player_data["MP"]) * 38, 2)


    # last 2 games points,rebounds,assists per 38

    if 'pts_per_38_last_2' not in player_data.columns:
    
        player_data['pts_per_38_last_2'] = player_data.groupby('player')['pts_per_38'].rolling(window=2).mean().reset_index(0, drop=True)

        # player_data['rbs_per_38_last_2']
        player_data['rbs_per_38_last_2'] = player_data.groupby('player')['trb_per_38'].rolling(window=2).mean().reset_index(0, drop=True)

        # player_data['ast_per_38_last_2']
        player_data['ast_per_38_last_2'] = player_data.groupby('player')['ast_per_38'].rolling(window=2).mean().reset_index(0, drop=True)



        # #last 5 games points,rebounds,assists per 38
        # player_data['pts_per_38_last_5']
        player_data['pts_per_38_last_5'] = player_data.groupby('player')['pts_per_38'].rolling(window=5).mean().reset_index(0, drop=True)

        # player_data['rbs_per_38_last_5']
        player_data['rbs_per_38_last_5'] = player_data.groupby('player')['trb_per_38'].rolling(window=5).mean().reset_index(0, drop=True)

        # player_data['ast_per_38_last_5']
        player_data['ast_per_38_last_5'] = player_data.groupby('player')['ast_per_38'].rolling(window=5).mean().reset_index(0, drop=True)

        
        #last 10 games points,rebounds,assists per 38
        player_data['pts_per_38_last_10']
        player_data['pts_per_38_last_10'] = player_data.groupby('player')['pts_per_38'].rolling(window=10).mean().reset_index(0, drop=True)

        # player_data['rbs_per_38_last_10']
        player_data['rbs_per_38_last_10'] = player_data.groupby('player')['trb_per_38'].rolling(window=10).mean().reset_index(0, drop=True)

        # player_data['ast_per_38_last_10']
        player_data['ast_per_38_last_10'] = player_data.groupby('player')['ast_per_38'].rolling(window=10).mean().reset_index(0, drop=True)


        # #last 25 games points,rebounds,assists per 38
        # player_data['pts_per_38_last_25']
        player_data['pts_per_38_last_25'] = player_data.groupby('player')['pts_per_38'].rolling(window=25).mean().reset_index(0, drop=True)

        # player_data['rbs_per_38_last_25']
        player_data['rbs_per_38_last_25'] = player_data.groupby('player')['trb_per_38'].rolling(window=25).mean().reset_index(0, drop=True)

        # player_data['ast_per_38_last_25']
        player_data['ast_per_38_last_25'] = player_data.groupby('player')['ast_per_38'].rolling(window=25).mean().reset_index(0, drop=True)


        # #last 50 games points,rebounds,assists per 38
        # player_data['pts_per_38_last_50']
        player_data['pts_per_38_last_50'] = player_data.groupby('player')['pts_per_38'].rolling(window=50).mean().reset_index(0, drop=True)

        # player_data['rbs_per_38_last_50']
        player_data['rbs_per_38_last_50'] = player_data.groupby('player')['trb_per_38'].rolling(window=50).mean().reset_index(0, drop=True)

        # player_data['ast_per_38_last_50']
        player_data['ast_per_38_last_50'] = player_data.groupby('player')['ast_per_38'].rolling(window=50).mean().reset_index(0, drop=True)


        # #last 100 games points,rebounds,assists per 38
        # player_data['pts_per_38_last_100']
        player_data['pts_per_38_last_100'] = player_data.groupby('player')['pts_per_38'].rolling(window=100).mean().reset_index(0, drop=True)

        # player_data['rbs_per_38_last_100']
        player_data['rbs_per_38_last_100'] = player_data.groupby('player')['trb_per_38'].rolling(window=100).mean().reset_index(0, drop=True)

        # player_data['ast_per_38_last_100']
        player_data['ast_per_38_last_100'] = player_data.groupby('player')['ast_per_38'].rolling(window=100).mean().reset_index(0, drop=True)

        # #last 250 games points,rebounds,assists per 38
        # player_data['pts_per_38_last_250']
        player_data['pts_per_38_last_250'] = player_data.groupby('player')['pts_per_38'].rolling(window=250).mean().reset_index(0, drop=True)

        # player_data['rbs_per_38_last_250']
        player_data['rbs_per_38_last_250'] = player_data.groupby('player')['trb_per_38'].rolling(window=250).mean().reset_index(0, drop=True)

        # player_data['ast_per_38_last_250']
        player_data['ast_per_38_last_250'] = player_data.groupby('player')['ast_per_38'].rolling(window=250).mean().reset_index(0, drop=True)


    #players shot attempts per game last 10,50 games
    player_data['shot_attempts_per_game_last_10'] = player_data.groupby('player')['FGA'].rolling(window=10).mean().reset_index(0, drop=True)

    player_data['shot_attempts_per_game_last_50'] = player_data.groupby('player')['FGA'].rolling(window=50).mean().reset_index(0, drop=True)

    #players true shooting percentage last 10, 50 games
    player_data['true_shooting_percentage_last_10'] = player_data.groupby('player')['TS%'].rolling(window=10).mean().reset_index(0, drop=True)

    player_data['true_shooting_percentage_last_50'] = player_data.groupby('player')['TS%'].rolling(window=50).mean().reset_index(0, drop=True)


    #players rebounding percentage last 10, 50 games
    player_data['rebounding_percentage_last_10'] = player_data.groupby('player')['TRB%'].rolling(window=10).mean().reset_index(0, drop=True)

    player_data['rebounding_percentage_last_50'] = player_data.groupby('player')['TRB%'].rolling(window=50).mean().reset_index(0, drop=True)

    #players assist percentage last 10, 50 games
    player_data['assist_percentage_last_10'] = player_data.groupby('player')['AST%'].rolling(window=10).mean().reset_index(0, drop=True)

    player_data['assist_percentage_last_50'] = player_data.groupby('player')['AST%'].rolling(window=50).mean().reset_index(0, drop=True)


    #players number of free throw attempts per game last 10, 50 games
    player_data['free_throw_attempts_per_game_last_10'] = player_data.groupby('player')['FTA'].rolling(window=10).mean().reset_index(0, drop=True)
    
    player_data['free_throw_attempts_per_game_last_50'] = player_data.groupby('player')['FTA'].rolling(window=50).mean().reset_index(0, drop=True)



    return player_data

```

```python
##column ordering and changing
player_data = column_ordering(player_data)
```

```python
cols_added = horizons(player_data)

# print(f'header of the data: {cols_added.head()}')

```

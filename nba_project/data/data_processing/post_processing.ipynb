```python
import pandas as pd


player_data = pd.read_csv("C:/Users/athen/Desktop/PersonalProjects/Learning/nba_project/data/data_scraping/data/players_ordered.csv")
player_data = player_data.replace('Not With Team', 0)
player_data = player_data.replace('Player Suspended', 0)
```

```python

def column_ordering(player_data):

    player_data = player_data.sort_values(by = ["player", "date"])

    return player_data


def convert_minutes(mp):
        try:
            parts = mp.split(":")
            if len(parts) == 2:
                return int(parts[0]) + int(parts[1]) / 60
            else:
                return float('nan')  # or handle it as needed
        except Exception as e:
            return float('nan')  # or handle it as needed


#feature engineering
def add_predictors(player_data):

    unique_players = player_data["player"].unique()

    
    cols_to_convert = ['PTS', 'TRB', 'AST',]

    for col in cols_to_convert:
        player_data[col] = pd.to_numeric(player_data[col], errors='coerce')


    #the MP col stores minutes played as a string, in the format MM:SS, convert to float, unless the value is 0, keep it as a 0
    if player_data["MP"].dtype == 'object':
        # Convert 'MP' column to numeric minutes
        player_data["MP"] = pd.to_numeric(round(player_data["MP"].str.split(":").str[0]) + pd.to_numeric(player_data["MP"].str.split(":").str[1]) / 60 ,2) 

    
        
    #add points per 38 minutes
    player_data["points_per_38"] = round((player_data["PTS"] / player_data["MP"]) * 38,2)

    # #add rebounds per 38 minutes
    player_data["rebounds_per_38"] = round((player_data["TRB"] / player_data["MP"]) * 38,2)

    # #add assists per 38 minutes
    player_data["assists_per_38"] = round(player_data["AST"] / player_data["MP"] * 38,2)


    #last 2 games points,rebounds,assists per 38
    player_data['pts_per_38_last_2']

    player_data['rbs_per_38_last_2']

    player_data['ast_per_38_last_2']

    #last 5 games points,rebounds,assists per 38
    player_data['pts_per_38_last_5']

    player_data['rbs_per_38_last_5']

    player_data['ast_per_38_last_5']

    #last 10 games points,rebounds,assists per 38
    player_data['pts_per_38_last_10']

    player_data['rbs_per_38_last_10']

    player_data['ast_per_38_last_10']

    #last 25 games points,rebounds,assists per 38
    player_data['pts_per_38_last_25']

    player_data['rbs_per_38_last_25']

    player_data['ast_per_38_last_25']

    #last 50 games points,rebounds,assists per 38
    player_data['pts_per_38_last_50']

    player_data['rbs_per_38_last_50']

    player_data['ast_per_38_last_50']

    #last 100 games points,rebounds,assists per 38
    player_data['pts_per_38_last_100']

    player_data['rbs_per_38_last_100']

    player_data['ast_per_38_last_100']

    #last 250 games points,rebounds,assists per 38
    player_data['pts_per_38_last_250']

    player_data['rbs_per_38_last_250']

    player_data['ast_per_38_last_250']


    #players shot attempts per game last 5, 25, 50, 100 games
    player_data['shot_attempts_per_game_last_10']

    player_data['shot_attempts_per_game_last_100']

    #players true shooting percentage last 10, 100 games
    player_data['true_shooting_percentage_last_10']

    player_data['true_shooting_percentage_last_100']


    #players rebounding percentage last 10, 100 games
    player_data['rebounding_percentage_last_10']

    player_data['rebounding_percentage_last_100']

    #players assist percentage last 10, 100 games
    player_data['assist_percentage_last_10']

    player_data['assist_percentage_last_100']


    #players number of free throw attempts per game last 10, 100 games
    player_data['free_throw_attempts_per_game_last_10']
    
    player_data['free_throw_attempts_per_game_last_100']



    #opponents defensive rating
    player_data['opp_defensive_rating']

    #opponents pace
    player_data['opp_pace']

    #opponents points allowed per game
    player_data['opp_points_allowed_per_game']

    #oppoonents total rebounds per game
    player_data['opp_total_rebounds_per_game']

    #opponents total assists allowed per game
    player_data['opp_total_assists_allowed_per_game']

    #opponents offensive rebounds allowed per game
    player_data['opp_offensive_rebounds_allowed_per_game']

    #opponents defensive rebounds allowed per game
    player_data['opp_defensive_rebounds_allowed_per_game']




    #team pace
    player_data['team_pace']

    #team points scored per game
    player_data['team_ppg']

    #team rebounds per game
    player_data['team_rpg']

    #team defensive rebounds per game
    player_data['team_defensive_rpg']

    #team offensive rebounds per game
    player_data['team_offensive_rpg']

    #team assists per game
    player_data['team_apg']


    #player usage rates
    player_data['usage_rate']

    #the superteam factor (overall team capability)
    player_data['superteam_factor']


    
    #days since last game
    player_data['days_since_last_game']

    #num of games in last 5 days
    player_data['num_games_last_5_days']

    #num of games in last 10 days
    player_data['num_games_last_10_days']

    #game importance
    player_data['regular_season']   

    player_data['playoffs']

    player_data['conf_finals']

    player_data['finals']


    return player_data


def combination_features(player_data):
    #combine features to create new features

    pass


def data_splitting(player_data):
    #split data into training, validation, and testing sets

    pass
```
